```python

results = []

for prompt in prompts:
    print(f"Generating images for prompt: {prompt}")

    folder_name = prompt[:40].replace(" ", "_").replace(",", "")  # Truncate + clean
    save_path = os.path.join(BASE_DIR, folder_name)
    os.makedirs(save_path, exist_ok=True)

    for i in range(IMAGES_PER_PROMPT):
        image = pipe(prompt).images[0]  # Generate image

        filename = f"{datetime.now().strftime('%Y%m%d_%H%M%S')}_{i}.png"
        image_path = os.path.join(save_path, filename)
        image.save(image_path)

        # Log to results
        results.append({
            "prompt": prompt,
            "image_path": image_path
        })

print("✅ All images generated!")
```
